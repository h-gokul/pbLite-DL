```python
# Data Augmentation
    
def AugmentImages(img):  
    # change color format
    # code = np.random.randint(0, 5) # more probability for image to be sent as it is
    # if code == 0:
    #     img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    # if code == 1:
    #     img = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    # if code == 2:
    #     img = cv2.cvtColor(im, cv2.COLOR_BGR2YCR_CB)
    # if code == 3:
    #     img = cv2.cvtColor(im, cv2.COLOR_BGR2LAB)
    # if code == 4:
    #     img = cv2.cvtColor(im, cv2.COLOR_BGR2LUV)

    # flip
    code = np.random.randint(-2, 2) # more probability for image to be sent as it is
    if code > -2:
        img = cv2.flip(im, code) 

    # change brightness, contrast
    """
    alpha 1  beta 0      --> no change  
    0 < alpha < 1        --> lower contrast  
    alpha > 1            --> higher contrast  
    -127 < beta < +127   --> good range for brightness values
    """
    beta = int(np.random.uniform(-50, 50))
    alpha = 1 + np.random.uniform(-0.2, 0.2)
    img = alpha*img + beta

    # change noise
    noise = np.zeros(img.shape, np.uint16)
    noise = cv2.randn(noise, 0, 15)
    img = cv2.add(img, noise, dtype=cv2.CV_8UC3)
    return img
img = AugmentImages(im)

plt.imshow(img)
```

```python
def basicUnit(model,filters, n_units,n_blocks):
    if (n_units == 1):   # to downsample on first layer of every block
        strides = (2, 2)
    else:
        strides = (1, 1)
        
    I = identityUnit(model,n_blocks) # skip connection is taken in the beginning, by default

    ##### Layer 1 ###########################################################################
    model = tf.layers.batch_normalization(model)
    model = tf.nn.relu(model)
    if ((n_units == 1)&(n_blocks == 1)) :  # get skip connection here only if it is the first unit of first block
        strides = (1,1)  # to avoid downsampling first unit of first block
        print('pre-activation skip connection accessed for 1st unit 1st block')
        I = identityUnit(model=model,n_blocks=1)
        
    model = tf.layers.conv2d(inputs = model, filters=filters,kernel_size=3,strides=strides, padding='SAME', activation = None)
    
    ##### Layer 2 ###########################################################################
    model = tf.layers.batch_normalization(model)
    model = tf.nn.relu(model)
    model = tf.layers.conv2d(inputs = model, filters=filters,kernel_size=3, strides = (1,1), padding='SAME', activation = None)
    
    model = tf.math.add(model, I)
    
    return model

def identityUnit(model,n_blocks):
    if n_blocks ==1:
        return model
    
    if n_blocks ==2:
        model = tf.nn.avg_pool(model, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME')
        return tf.pad(model, [[0, 0], [0, 0], [0, 0], [8,8]])
    
    if n_blocks ==3:
        model = tf.nn.avg_pool(model, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME')
        return tf.pad(model, [[0, 0], [0, 0], [0, 0], [16,16]])


def ResNet(Img,num_classes=10):
    
    model = tf.layers.conv2d(inputs = Img, filters=16,kernel_size=3, padding='SAME', activation = None)
    
    model = resBlock(model,filters=16,n_blocks=1)
    model = resBlock(model,filters=32,n_blocks=2)
    model = resBlock(model,filters=64,n_blocks=3)
        
    model = tf.layers.batch_normalization(model)
    model = tf.nn.relu(model)
    
    model = tf.layers.flatten(model)
    
    model = tf.layers.dense(inputs = model, units = 128, activation = tf.nn.relu)

    prLogits = tf.layers.dense(inputs = model,units=num_classes, activation=None)
    prSoftMax = tf.nn.softmax(prLogits, axis=None, name=None)
    
    return prLogits, prSoftMax

def resBlock(model,filters,n_blocks):
    print( '\n ResNet Block: '+ str(n_block),' \t No. of filters: '+ str(filters))
    
    for i in range(1,4): # 3 units per block
        print( '\n \t ResNet Unit: '+ str(i))
        tmp = basicUnit(model= model, filters=filters, n_units = i, n_blocks = n_blocks)
        model=tmp
    return model

# Reference : https://github.com/chao-ji/tf-resnet-cifar10/blob/master/files/ResNet20_CIFAR10.pdf
    

    
    
```

```python
# DenseNet

config = tf.ConfigProto()
config.gpu_options.allow_growth = True

print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))
```

```python
# https://www.kaggle.com/genesis16/densenet-93-accuracy

def DenseBlock(model, n_filters = 36, dropout = 0.2, l=12, compression = 0.5):
    tmp = model
    
    for _ in range(l):
        model = tf.layers.batch_normalization(tmp)
        model = tf.nn.relu(model)
        model = tf.layers.conv2d(model, filters=int(n_filters*compression),kernel_size=3, padding='SAME', activation = None)

        if dropout>0:
            model = tf.nn.dropout(model, dropout)
        
        tmp = tf.concat([tmp,model],axis=-1)         
    return tmp

def TransitionBlock(model, n_filters = 12, dropout = 0.2, compression = 0.5):
    model = tf.layers.batch_normalization(model)
    model = tf.nn.relu(model)
    model = tf.layers.conv2d(model, filters=int(n_filters*compression),kernel_size=1, padding='SAME', activation = None)

    if dropout>0:
        model = tf.nn.dropout(model, dropout)
    
    model = tf.layers.AveragePooling2D(model,pool_size=(2,2))
    return model

def DenseNet(Img, n_filters = 36, dropout = 0.2, l= 12,compression = 0.5):
    num_classes = 10
    
    model = tf.layers.conv2d(inputs = Img, filters=36,kernel_size=3, padding='SAME', activation = None)

    # Dense + Transition Block : 1
    model = DenseBlock(model, n_filters, dropout)
    model = TransitionBlock(model, n_filters, dropout)

    # Dense + Transition Block : 2
    model = DenseBlock(model, n_filters, dropout)
    model = TransitionBlock(model, n_filters, dropout)

    # Dense + Transition Block : 3
    model = DenseBlock(model, n_filters, dropout)
    model = TransitionBlock(model, n_filters, dropout)

    # Dense + Transition Block : 4
    model = DenseBlock(model, n_filters, dropout)

    # output stage
    model = tf.layers.batch_normalization(model)
    model = tf.nn.relu(model)
    model = tf.layers.AveragePooling2D(pool_size=(2,2))(model)
    model = tf.layers.flatten(model)
    prLogits = tf.layers.dense(inputs = model,units=10, activation=None)
    
    prSoftMax = tf.nn.softmax(prLogits, axis=None, name=None)

    return prLogits, prSoftMax

# Hyperparameters

num_classes = 10
l = 12
n_filters = 36 #added 24 more filters
compression = 0.5 
dropout_rate = 0.2


prLogits, prSoftMax = DenseNet(Img )
```

```python

def ResBlock(model, in_channel, out_channel, n, pool_stride, cardinality,base_width):
    I = model
    for i in range(n):
        if i == 0:
            tmp = BottleNeck(model,I,in_channel, out_channel,pool_stride, cardinality,base_width) # stride is predefined  for first block
        else: 
            tmp = BottleNeck(model,I,out_channel, out_channel,1, cardinality,base_width) # same stride =1 for rest of other blocks
        model = tmp
    return model
            
def BottleNeck(model,I, in_channels, out_channels,pool_stride, cardinality,base_width):
    
    D = cardinality * int(base_width//64)

    if in_channels!=out_channels:
        I = tf.layers.conv2d(inputs = I, filters=outchannels,kernel_size=1, stride = 1, padding='SAME', activation = None) 
        I = tf.layers.batch_normalization(I)
    
    model = tf.layers.conv2d(inputs = model, filters=D,kernel_size=1, stride = 1, padding='SAME', activation = None)
    model = tf.layers.batch_normalization(model)
    model = tf.nn.relu(model)
    
    model = tf.layers.conv2d(model, filters = D,kernel_size=3, stride = pool_stride, padding='SAME', activation = None)
    model = tf.layers.batch_normalization(model)
    model = tf.nn.relu(model)
    
    model = tf.layers.conv2d(model, filters = out_channels,kernel_size=1, stride = 1, padding='SAME', activation = None)
    model = tf.layers.batch_normalization(model)

    model = tf.math.add(model, I)

    model = tf.nn.relu(model)
    return model    

def ResNext(Img, depth = 20,base_width = 32, cardinality=3):
    n = (depth-2)//9 # block depth is 2
    
    # first layers
    model = tf.layers.conv2d(inputs = Img, filters=32,kernel_size=3, padding='SAME', activation = None)
    model = tf.layers.batch_normalization(model)
    model = tf.nn.relu(model)

    model = ResBlock(model,in_channel =16, out_channel = 32, n=n, pool_stride = 1, cardinality = cardinality, base_width = base_width)
    model = ResBlock(model,in_channel =32, out_channel = 64, n=n,  pool_stride = 2, cardinality = cardinality, base_width = base_width)
#     model = ResBlock(model,in_channel =64, out_channel = 128, n,  pool_stride=2,cardinality= cardinality, base_width = base_width)

    model = tf.layers.AveragePooling2D(8,1)(model)

    model = tf.layers.flatten(model)
    model = tf.layers.dense(inputs = model,units=256, activation=None)
    prLogits = tf.layers.dense(model,units=10, activation=None)

    prSoftMax = tf.nn.softmax(prLogits, axis=None, name=None)

    return prLogits, prSoftMax
```

```python

def bottleNeckUnit(model): 
    model = tf.layers.conv2d(inputs = Img, filters=32,kernel_size=1, padding='SAME', activation = None)
    model = tf.layers.batch_normalization(model)
    model = tf.nn.relu(model)
    
    model = tf.layers.conv2d(model, filters=64,kernel_size=3, padding='SAME', activation = None)
    model = tf.layers.batch_normalization(model)
    model = tf.nn.relu(model)
    
    model = tf.layers.conv2d(model, filters=32,kernel_size=1, padding='SAME', activation = None)
    model = tf.layers.batch_normalization(model)
    model = tf.nn.relu(model)

    return model

def merge(C_splits):
    model = C_splits[0]
    for i in range(1, len(C_splits)):
        model = tf.math.add(model,C_splits[i])
    return model

def splitUnit(model, stride,cardinality):
    C_splits = []
    for i in range(cardinality) :
        split = bottleNeckUnit(model)
        C_splits.append(split)

    model = merge(C_splits)
    return model 

def ResBlock(model):
    # split + transform(bottleneck) + transition + merge
    I =  model
    model = splitUnit(model)
    model = tf.add(model,I)
    return model 

def ResNext(Img, depth = 20,base_width = 32, cardinality=3):

    # first layers
    model = tf.layers.conv2d(inputs = Img, filters=32,kernel_size=3, padding='SAME', activation = None)
    model = tf.layers.batch_normalization(model)
    model = tf.nn.relu(model)

    model = ResBlock(model)
    model = ResBlock(model)

    model = tf.layers.flatten(model)
    model = tf.layers.dense(inputs = model,units=256, activation=None)
    prLogits = tf.layers.dense(model,units=10, activation=None)

    prSoftMax = tf.nn.softmax(prLogits, axis=None, name=None)

    return prLogits, prSoftMax
```
